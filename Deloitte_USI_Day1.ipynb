```python
#importing the essential libraries
from openai import OpenAI #through OpenAI we configure the OpenAI API as well as the LLM
import os #for creating a virtual env
import json #for getting structured output
```

```python
#configuring the OpenAI API
os.environ["OPENAI_API_KEY"]="YOUR API KEY"
client=OpenAI(
    api_key=os.environ["OPENAI_API_KEY"]
)
```

```python
#configuring the LLM
response=client.responses.create(
    model="gpt-4.1-mini", #model chosen from the GPT Family of Models
    input="Explain Generative AI as if I were a school student" #user query
)

print(response.output_text)
```

Undestanding LLM Settings or Model Parameters:

- temperature

```python
response=client.responses.create(
    model="gpt-4.1-mini", #model chosen from the GPT Family of Models
    input="Explain Generative AI as if I were a school student",#user query
    temperature=0.2 #low temperature makes the output more deterministic and consistent in nature
)

print(response.output_text)
```

1. Zero Shot Prompting - Retail Client has collected customer feedback from multiple channels. Consultants need to identify recurring business themes before recommending improvements

- Delivery was delayed by 4 days
- Product quality was excellent
- Mobile application crashes frequently
- Pricing is higher as compared to the competitors
- Customer Support was polite but took very long to respond

```python
zero_shot_prompt="""You are a Management Consultant,
Analyze the customer feedback below and perform the following tasks:

1. Identify recurring themes
2. Group similar feedback together
3. Assign a business priority
4. Suggest business recommendation for each theme.

Return the results in a markdown table

Customer Feedback:
- Delivery was delayed by 4 days
- Product quality was excellent
- Mobile application crashes frequently
- Pricing is higher as compared to the competitors
- Customer Support was polite but took very long to respond
"""
```

```python
response=client.responses.create(
    model="gpt-4.1-mini", #model chosen from the GPT Family of Models
    input=zero_shot_prompt,#user query
)

print(response.output_text)
```

2. Few Shot Prompting - A client is willing to monitor customer sentiment across different feedback channels

```python
few_shot_prompt="""
You are a Customer Experience Consultant. You are supposed to classify each feedback into:

- Positive
- Negative
- Neutral

Examples:

Feedback - "The delivery was quick and the packaging was excellent"
Output Sentiment - Positive
Confidence - 97%

Feedback - "The product quality is ok but packaging could be improved"
Output Sentiment - Neutral
Confidence - 60%

Now classify - "The support executive transferred the call 3 times and my issue is still unresolved"
"""
```

```python
response=client.responses.create(
    model="gpt-4.1-mini", #model chosen from the GPT Family of Models
    input=few_shot_prompt,#user query
)

print(response.output_text)
```
